```python
import torch
import torch.nn.functional as F
from matplotlib import pyplot as plt
```

# ReLU

```python
x = torch.linspace(-5, 5, 200)   # 构造一段连续的数据
x_np = x.data.numpy()    #plt中形式需要numoy形式，tensor形式会报错

y_relu = F.relu(x).data.numpy()
plt.plot(x_np, y_relu, c='red', label='relu')
# plt.ylim((-1, 5))
plt.legend(loc='best')

plt.show()
```

```python
x.requires_grad = True
y_relu = F.relu(x)
z = torch.sum(y_relu)
z.backward()

plt.plot(x_np, x.grad.data.numpy(), c='red', label='Relu Derivative')
# plt.ylim((-0.01, 1.2))
plt.legend(loc='best')

plt.show()
```

# Sigmoid

```python
y_relu = torch.sigmoid(x).data.numpy()
plt.plot(x_np, y_relu, c='red', label='Sigmoid')
# plt.ylim((-0.1, 1.2))
plt.legend(loc='best')

plt.show()
```
